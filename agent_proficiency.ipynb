```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Agent Name": ["Agent 1", "Agent 2", "Agent 3", "Agent 4", "Agent 5"],
    "Customer Rating": [4.5, 4.2, 4.8, 4.3, 4.1],
    "Number of Customers": [120, 150, 130, 110, 140],
    "Resolution Time": [15, 18, 14, 20, 17]
}

df = pd.DataFrame(data)

df.to_csv("agent_proficiency.csv", index=False)

data = pd.read_csv("agent_proficiency.csv")

def calculate_proficiency(data):
    data["Proficiency Score"] = (data["Customer Rating"] * data["Number of Customers"]) / data["Resolution Time"]
    return data

proficiency_data = calculate_proficiency(data)

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(proficiency_data["Agent Name"], proficiency_data["Proficiency Score"], color='skyblue')
ax.set_xlabel("Agent Name")
ax.set_ylabel("Proficiency Score")
ax.set_title("Agent Proficiency")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(proficiency_data["Number of Customers"], labels=proficiency_data["Agent Name"], autopct='%1.1f%%', colors=['blue', 'orange', 'green', 'red', 'purple'])
ax.set_title("Customer Distribution Among Agents")
plt.show()

print(proficiency_data)
```
